## Protokół HTTP

HyperText Transfer Protocol

**Metody HTTP:**
- GET - pobierz zasoby (np. pobranie zawartości strony internetowej)
- POST - wyślij informację (i ewentualnie odbierz informację zwrotną) np. wysłanie formularza
- PUT - utwórz nowy zasób
- DELETE - usuń zasoby
- itd. ...

**Odpowiedź (Response):**
- zawiera tekst odpowiedzi
- zawiera kod odpowiedzi -> https://pl.wikipedia.org/wiki/Kod_odpowiedzi_HTTP
- dodatkowe informacje
